```python
import pandas as pd
data = pd.read_csv('Africa_climate_change.csv')
```

```python
data.head(10)
```

```python
data.tail(10)
```

```python
data.info()
```

```python
data.drop(columns=["PRCP", "TMAX", "TMIN"],inplace=True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

```python
data.head(10)
```

```python
data.tail(10)
```

```python
import datetime
data["DATE"]=pd.to_datetime(data["DATE"])
```

```python
data.info()
```

```python
data["YEAR"]=data["DATE"].dt.year
data["MONTH"]=data["DATE"].dt.month
```

```python
data.info()
```

```python
data
```

```python
df = data[(data["COUNTRY"] =="Cameroon") | (data["COUNTRY"] =="Tunisia")]
df
```

```python
df.isnull().sum()
```

```python
df1 = df.groupby(["COUNTRY", "DATE"])["TAVG"].mean()
```

```python
df1.isnull().sum()
```

```python
df1.tail(50)
```

```python
df1 = df1.reset_index()
df1
```

```python
import plotly.express as px
```

```python
fig = px.line(df1, x="DATE", y='TAVG', color = "COUNTRY", title='Africa_climate_change')
fig.show()
```

```python
df2 = df.groupby(["COUNTRY", "YEAR"])["TAVG"].mean()
```

```python
df2.isnull().sum()
```

```python
df2.tail(50)
```

```python
df2 = df2.reset_index()
df2
```

```python
fig = px.line(df2, x="YEAR", y='TAVG', color = "COUNTRY", title='Africa_climate_change')
fig.show()
```

```python
df3 = df.groupby(["COUNTRY", "MONTH"])["TAVG"].mean()
```

```python
df3.isnull().sum()
```

```python
df3.info()
```

```python
df3=df3.reset_index()
df3
```

```python
df3.head(10)
```

```python
fig = px.line(df3, x="MONTH", y='TAVG', color = "COUNTRY", title='Africa_climate_change')
fig.show()
```
