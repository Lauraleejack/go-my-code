```python


from ydata_profiling import ProfileReport

profile=ProfileReport(data,title='EDUCATION_ATTAINMENT',minimal=True)
profile.to_notebook_iframe()
```

```python
profile.to_html()
```

the dataset shows missing values for both males and females. the data which is not recorded has no paticular pattern and are rather scattered. There is a relationship between the variables and the missing data.


The report provides correlation matrices to visualize the strength and direction of relationships between different variables. Some strong correlations exist on the age variable.

Strong positive correlation between educational attainment rates for the entire population, males, and females. This suggests that countries with higher overall educational attainment rates also have higher rates of male and female educational attainment.

Wide range of educational attainment rates across countries. This suggests that there is a significant gap in educational attainment between countries.

Higher educational attainment rates for males than females. The percentage of males that has grades  is consistently higher than the percentage of females. This suggests that there is a gender gap in educational attainment.
